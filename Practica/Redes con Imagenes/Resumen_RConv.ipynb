## Consejo para el manejo de Imagenes

Las imagenes es una composicion de pixeles que tienen un rango numero de 0(negro) a 255(blanco) . 
Para una escala de grises el 0 es negro y 1 es blanco y todos los numeros que estan en la mitad es la escala de grises.

Las imagenes tienen un ancho y un largo que es el tamano de la imagen, por dentro de cada recuadro es la cantidad de pixeles.

La maquina interpreta una matriz, la maquina no va ver un 8 sino va a ver  una composicion de numero de la matriz que si le metemos color o pasamos una imagen que el humano va a interteprar una imagen pero la maquina ve valores numericos.

![CNN](https://static.platzi.com/media/user_upload/1_v1LM663dEdoPImnefa2zBQ-1c988aab-a29a-4a53-972c-3185daf1a0e8.jpg)

Cuando trabajamos con imagenes RGB en redes neuronales, es recomendable trabajar con un canal (escala de grises) que trabajar con RGB siempre cuando los colores no sean impotantes.

Hay casos en que las imagenes es recomendable trabajar con RGB ya que los colores son importante para resaltar algunos de los canales.

Otra recomendacion, es trabajar con una escala de dimension definida para todas las imagenes, ya que la red neuronal entrena con imagenes nxn de una capa debo enviarle exactamente con esa info a la red neuronal para que pueda realizar la prediccion.
Es importante trbajar con las mismas de dimensiones ya sea para precision y entrenamiento.

## El Kernel (Filtro)

El kernel es simplemente esta matriz:

$[[1, 1, 1],$
$[0, 0, 0],$
$[-1, -1, -1]]$

Esta matriz se desliza desde la esquina superior izquierda hasta la esquina inferior derecha paso por paso hasta terminar toda la imagen haciendo una operacion matematica.

Esta operacion se llama **CONVOLUCION**

## Capa de pooling

El problema: Multiples filtros en cada capa hacen una red compleja con muchos parámetros.

MaxPooling : Opera pixel por pixel y saco el maximo valor del kernel que sería la salida.

Average: Opera pixel por pixel y saco el promedio de ese kernel que sería la salida.

No estamos perdiendo informacion, sino sacando ruido.

Es mas comun usar el MaxPooling

## Arquitectura de Redes Convolucionales

Tengo una o multiples imagenes de entradas, esas imagenes van a pasar por una red de convolucion donde se va a extraer caracteristicas como bordes, texturas, mucho mas superficiales pero cuando voy entrando mas profundo a la red . Esas capas de convolucion estan entrelazadas a una capa de MaxPooling de lo cual va a reducir la complejidad acortando el largo y ancho de la imagen y asi capas, capas de convoluciones con maxPooling nos van a devolver imagenes mas pequeñas pero con mayor profundidad en su contexto. Despues voy aplicar una capa de Flatten en donde me va a llevar el arreglo o tensor resultante a una dimension. 
Con esta dimension puedo hacer la clasificacion con capas Densa.

Entonces:

- Convolucion: aumenta la profundidad
- Pooling: reduce el largo y ancho

- Flatten: lleva a una dimension para mas que la clasificacion sea mas facil sin perder informacion

- Densas: lo uso para aplicar la clasificacion

Para calcular las dimensiones de la red luego de aplicar la convolución se usa la siguiente formula:
![CNN](https://static.platzi.com/media/user_upload/cnn-ecd384a2-d3c4-4975-84ff-74bfca495962.jpg)

Por ejemplo para una imagen con medidas iniciales de 28x28x1 que tiene 1 filtro de 3x3 , un padding de “same” , stride 1 y 16 canales de salida se tiene lo siguiente:
n eje x= (28+2(1) -3)/1 +1 =28

n eje y= (28+2(1) -3)/1 +1 =28

Siendo las dimensiones de salida 28x28x16
